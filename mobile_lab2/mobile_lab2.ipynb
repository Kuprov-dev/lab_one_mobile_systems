```python
data=data.iloc[:,0:17]
```

```python
data['t_first']=pd.to_datetime(data['t_first'])
```

```python
time_series=data.set_index('t_first')
```

```python
time_series=time_series.resample('T').sum().fillna(0)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,10))
plt.plot(time_series['in_bytes'])
plt.show
```
